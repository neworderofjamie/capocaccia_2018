```python
%matplotlib inline

import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.feature import hog
from skimage import exposure
from glob import glob

# Make figures big enough to see
plt.rcParams["figure.figsize"] = (16, 16)

# Snapshot dimensions
width = 450
height = 50

# Step between images with which to compare
step = 25

def load_route(path):
    # Use glob to find images in path and then sort by INTEGER number
    wildcard = path + "/image_*.png"
    filenames = sorted(glob(wildcard), key=lambda f: int(f[len(path) + 7:len(f) - 4]))
    
    # Read images
    data = np.empty((height, width, len(filenames)))
    for i, f in enumerate(filenames):
        data[:,:,i] = cv2.imread(f, cv2.IMREAD_GRAYSCALE)
    
    # Return data, throwing away first frame cos it's always blank
    return data[:,:,1:]

def load_spherical(path):
     # Use glob to find images in path and then sort by INTEGER number
    wildcard = path + "/image_*.csv"
    filenames = sorted(glob(wildcard), key=lambda f: int(f[len(path) + 7:len(f) - 4]))
    
    # Read images
    data = np.empty((36, len(filenames)))
    for i, f in enumerate(filenames):
        data[:,i] = np.loadtxt(f, delimiter=",", skiprows=1, usecols=[2])
    
    # Return data, throwing away first frame cos it's always blank and first coefficient as it's DC
    return data[1:,1:]

def calc_idf(image, images):
    # Vector to hold differences
    differences = np.empty(images.shape[2])
    
    # Loop through images in array, calculate difference and add to array
    for i in range(images.shape[2]):
        differences[i] = np.sqrt(np.sum((image - images[:,:,i]) ** 2))
    
    return differences

def plot_comparison(data1, data2, step=step):
    # Indices of subset of images from 1st array to compare
    test_indices = range(0, data1.shape[2], step)

    # Create axes to show differences
    fig, axes = plt.subplots(len(test_indices), 3, sharex="col")

    # Loop through test indices
    for i, x in enumerate(test_indices):
        axes[i, 0].set_title("Position %u" % x)
        axes[i, 0].set_ylabel("RMS image difference")

        # Compare test image from 1st image with each of second images
        comp_data2 = calc_idf(data1[:,:,x], data2)
        best_data2_index = np.argmin(comp_data2)

        # Plot the IDF and overlay vertical line with best match
        axes[i, 0].plot(comp_data2)
        axes[i, 0].axvline(best_data2_index, color="red", linestyle="--")

        # Show image from 1st array and best match from 2nd
        axes[i, 1].imshow(data1[:,:,x], cmap="gray")
        axes[i, 2].imshow(data2[:,:,best_data2_index], cmap="gray")

    axes[-1, 0].set_xlabel("Data 1 index")

def calc_hog(images):
    hog_images = np.empty(((width * height * 8) / (10 * 10), images.shape[2]))
    for i in range(images.shape[2]):
        hog_images[:,i] = hog(images[:,:,i], orientations=8, pixels_per_cell=(10, 10),
                              cells_per_block=(1, 1), visualise=False)
    return hog_images
 
def calc_hog_idf(image, images):
    # Vector to hold differences
    differences = np.empty(images.shape[1])
    
    # Loop through images in array, calculate difference and add to array
    for i in range(images.shape[1]):
        differences[i] = np.sqrt(np.sum((image - images[:,i]) ** 2))
    
    return differences

def plot_hog_comparison(hog1, hog2, data1, data2, step=step):
    # Indices of subset of images from 1st array to compare
    test_indices = range(0, hog1.shape[1], step)

    # Create axes to show differences
    fig, axes = plt.subplots(len(test_indices), 3, sharex="col")

    # Loop through test indices
    for i, x in enumerate(test_indices):
        axes[i, 0].set_title("Position %u" % x)
        axes[i, 0].set_ylabel("HOG image difference")

        # Compare test image from 1st image with each of second images
        comp_hog2 = calc_hog_idf(hog1[:,x], hog2)
        best_hog2_index = np.argmin(comp_hog2)

        # Plot the IDF and overlay vertical line with best match
        axes[i, 0].plot(comp_hog2)
        axes[i, 0].axvline(best_hog2_index, color="red", linestyle="--")

        # Show image from 1st array and best match from 2nd
        axes[i, 1].imshow(data1[:,:,x], cmap="gray")
        axes[i, 2].imshow(data2[:,:,best_hog2_index], cmap="gray")

    axes[-1, 0].set_xlabel("Data 1 index")
```

# Restaurant

## Raw IDF

```python
# Load two arrays of routes
data1 = load_route("drive")
data2 = load_route("drive2")
plot_comparison(data1, data2)
```

## Spherical harmonic

```python
spherical1 = load_spherical("drive")
spherical2 = load_spherical("drive2")
plot_hog_comparison(spherical1, spherical2, data1, data2)
```

## HOG

```python
hog1 = calc_hog(data1)
hog2 = calc_hog(data2)

plot_hog_comparison(hog1, hog2, data1, data2)
```
